# Concatenating All the csvs

```python
csv_file_list = ['GWStation-96', 'GWStation-97', 'GWStation-98', 'GWStation-99', 'GWStation-00',
                'GWStation-01', 'GWStation-02', 'GWStation-03', 'GWStation-04', 'GWStation-05',
                'GWStation-06', 'GWStation-07', 'GWStation-08', 'GWStation-09', 'GWStation-10',
                'GWStation-11', 'GWStation-12', 'GWStation-13', 'GWStation-14', 'GWStation-15',
                'GWStation-16', 'GWStation-17', 'GWStation-18']

list_of_dataframes = []
for filename in csv_file_list:
    list_of_dataframes.append(pd.read_csv(filename))

df = pd.concat(list_of_dataframes)
```

```python
df = df[['STATE', 'DISTRICT', 'STATION', 'LATITUDE', 'LONGITUDE', 'STATION_TYPE', 'MAY (Level m)', 'AUG (Level m)', 'YEAR']]
```

```python
import geopandas as gpd
```

```python
dfGeo = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.LONGITUDE, df.LATITUDE))
```

```python
dfGeo.head()
```

```python
dfGeo96 = dfGeo[dfGeo.YEAR == 1996]
dfGeo18 = dfGeo[dfGeo.YEAR == 2018]
```

```python
import altair as alt
```

```python
alt.data_transformers.disable_max_rows()
```

```python
map96 = alt.Chart(dfGeo96).mark_geoshape().encode(
    color=alt.Color('MAY (Level m):Q', title='MAY (Level m)', scale=alt.Scale(scheme='inferno', domain=[55,0])),
    tooltip=['MAY (Level m):Q','STATE','DISTRICT','STATION', 'YEAR'],
).properties(
    title='Station-wise Groudnwater Depth 1996',
    width=500,
    height=500
).configure_view(
    strokeWidth=0
)

map18 = alt.Chart(dfGeo18).mark_geoshape().encode(
    color=alt.Color('MAY (Level m):Q', title='MAY (Level m)', scale=alt.Scale(scheme='inferno', domain=[55,0])),
    tooltip=['MAY (Level m):Q','STATE','DISTRICT','STATION', 'YEAR'],
).properties(
    title='Station-wise Groudnwater Depth 2018',
    width=500,
    height=500
).configure_view(
    strokeWidth=0
)
```
